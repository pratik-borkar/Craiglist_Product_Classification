```python
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import re

# Download necessary NLTK data
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to convert NLTK tag to WordNet tag
def nltk_tag_to_wordnet_tag(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:          
        return None

# Function to lemmatize a sentence
def lemmatize_sentence(sentence):
    # Tokenize the sentence
    nltk_tagged = nltk.pos_tag(nltk.word_tokenize(sentence))  
    # Convert to wordnet's format
    wordnet_tagged = map(lambda x: (x[0], nltk_tag_to_wordnet_tag(x[1])), nltk_tagged)
    lemmatized_sentence = []
    for word, tag in wordnet_tagged:
        if tag is None:
            # if there is no available tag, append the token as is
            lemmatized_sentence.append(word)
        else:        
            # else use the tag to lemmatize the token
            lemmatized_sentence.append(lemmatizer.lemmatize(word, tag))
    return " ".join(lemmatized_sentence)
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

from nltk.corpus import wordnet

# Function to map NLTK's POS tags to the format WordNetLemmatizer expects
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

def clean_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    # Lemmatize and remove stopwords
    lemmatized_tokens = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in tokens if word not in stop_words]
    # Join the tokens back into a string
    return ' '.join(lemmatized_tokens)
```

```python
# Convert columns to string and handle missing values
main_df['Name'] = main_df['Name'].fillna('').astype(str).apply(clean_text)
main_df['Description'] = main_df['Description'].fillna('').astype(str).apply(clean_text)

# Now apply your text processing functions
main_df['Name'] = main_df['Name'].apply(lambda x: lemmatize_sentence(x.lower()))
main_df['Description'] = main_df['Description'].apply(lambda x: lemmatize_sentence(x.lower()))

# creating new column information which is concatenation of product Name and Description, which stores overall context about any product
main_df["Information"] = main_df["Name"] + main_df["Description"]
main_df = main_df.drop(main_df.columns[2:4], axis=1)
main_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(main_df['Information'], main_df[main_df.columns[2:-1]], test_size=0.2, random_state=0, shuffle=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(strip_accents='unicode', analyzer='word', ngram_range=(1,3), norm='l2') # Using a tf-idf weighting scheme rather than normal boolean weights for better performance
vectorizer.fit(X_train) # Reference : https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html

X_train_transformed = vectorizer.transform(X_train)
X_test_transformed = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming y_train and y_test are already in the correct multi-label format

# Using Logistic Regression with balanced class weights
lr = LogisticRegression(solver='lbfgs', max_iter=1000, class_weight='balanced', random_state = 0)

# Wrap Logistic Regression in OneVsRestClassifier
model = OneVsRestClassifier(lr)

# Fit the model on the training data
model.fit(X_train_transformed, y_train)

# Get class predictions on the test data
y_pred = model.predict(X_test_transformed)

# Evaluate the model using multi-label classification metrics
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Initialize logistic regression models with different configurations
log_reg1 = OneVsRestClassifier(LogisticRegression(solver='lbfgs', max_iter=1000, C=1.0, class_weight='balanced', random_state = 0))
log_reg2 = OneVsRestClassifier(LogisticRegression(solver='liblinear', max_iter=5000, C=1.0, penalty='l2', random_state = 0))
gradient_boosting = OneVsRestClassifier(GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0))

# Fit models on the training data
log_reg1.fit(X_train_transformed, y_train)
log_reg2.fit(X_train_transformed, y_train)
gradient_boosting.fit(X_train_transformed, y_train)

# Predict probabilities for each class with each model
prob1 = log_reg1.predict_proba(X_test_transformed)
prob2 = log_reg2.predict_proba(X_test_transformed)
prob3 = gradient_boosting.predict_proba(X_test_transformed)

# Average the probabilities from all models
avg_probs = np.mean([prob1, prob2, prob3], axis=0)

# Convert average probabilities to final predictions (You can adjust the threshold as needed)
y_pred = (avg_probs >= 0.5).astype(int)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```
